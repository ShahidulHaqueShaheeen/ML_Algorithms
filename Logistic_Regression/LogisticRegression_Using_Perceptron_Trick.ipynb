```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1],c = y,cmap = 'winter',s = 50)
```

```python
X
```

```python
def decission(x):
    
    return 1 if x > 0 else 0
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1, axis = 1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        
        j = np.random.randint(0,100)
        y_hat = decission(np.dot(X[j],weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]
        
    return weights[0],weights[1:]
        
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor = LogisticRegression()
lor.fit(X,y)
```

```python
lor.coef_
```

```python
lor.intercept_
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

# So, we can say that, the logistic Regression is perform better as it goes untill it finds the perfect position whereas perceptron trick reamin stop untill it finds any misclassified point.
