```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features= 1, n_informative=1, n_targets=1, bias=0.0,  noise=80, random_state= 13)
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.show()
```

```python
# let's apply OLS
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X), color = 'red')
plt.show()
```

```python
# Let's apply GD and we have m = 78.35 and b = 0

y_predict = ((78.5 * X) + 0).reshape(4) # 2d to 1d
y_predict
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X), color = 'red', label = 'ols')
plt.plot(X,y_predict,color = 'green', label = 'b = 0')
plt.legend()
plt.show()
```
